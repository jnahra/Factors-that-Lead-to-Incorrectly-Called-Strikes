```python
def two_strike_foul(df):
    if (df['strikes'] == 2) and (df['description'] == 'foul'):
        return 'N'
    else:
        return df['type']
```

```python
df['new_type'] = df.apply(two_strike_foul,axis=1)
```

```python
df['new_type'].value_counts()
```

```python
df = df.loc[(df['new_type'] == 'B') | (df['new_type'] == 'S')]
```

```python
(df.groupby(by=['on_base','outs_when_up','count','new_type'])['delta_run_exp'].std()==0).value_counts()
```
